```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

original_image = cv2.imread('/content/low_contrast1.jpg')

min_val = np.min(original_image)
max_val = np.max(original_image)

new_min = 0
new_max = 255

stretched_image = ((original_image.astype(float) - min_val) * (new_max - new_min) / (max_val - min_val)) + new_min

stretched_image = np.clip(stretched_image, new_min, new_max).astype(np.uint8)

plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(original_image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(stretched_image)
plt.title('Stretched Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
